```python
def flip_n_times( times ):
    score = 0
    
    scoreArray = np.random.randint( 2, size=10) 
    score = sum(scoreArray)
    return score
```

```python
def flip_n_coins( num, times ):
    distribution = []
    for i in range( num ):
        score = flip_n_times( times )
        distribution.append( score )
    return distribution
```

```python
def get_v_min( num, times ):
    distribution = flip_n_coins( num, times )
    print(distribution)
    return min(distribution)
```

```python
def repeat_exp( repeat_time ):
    
    num = 1000
    times = 10
    
    sum_min = 0
    for i in range( repeat_time ):
        sum_min += get_v_min( num , times )

    avg_min = sum_min / repeat_time
    
    return avg_min
```

```python
v_min = repeat_exp(100 )
```

```python
class Point:
    x = 0
    y = 0 
    def __init__(self,x,y):
        self.x = x
        self.y = y
```

```python
class Line:
    p1 = None
    p2 = None
    y_int = 0
    slope = 0
    
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        
        dx = p2.x - p1.x
        dy = p2.y - p1.y
    
        # check if the line is vertical
        if dx == 0 :
            self.is_vertical = True
            self.slope = None 
        
        else:
            # calculate slope
            self.is_vertical = False
            self.slope = dy/dx
        
        # derive the formula for the line in terms of x an y
        # (y - y1) = m(x - x1) where (x1,y1) is coord of a point and m is the slope
        # General Form y = mx + b, we know m so only b is needed
        # we call it y_int since it is the y interception
        
        self.y_int = p1.y - self.slope * p1.x
          
    
    def mapping(self, p):
        # to check which side is p at and thus decide its sign
        
        # if line is not vertical, compare the y values
        if self.is_vertical == False:
            line_y_value = self.slope*p.x + self.y_int
            diff = p.y - line_y_value
            
        # if line is vertical, compare the x values
        else:
            line_x_value = self.p1.x
            diff = p.x - line_x_value
        
        return np.sign(diff)        
            
```
