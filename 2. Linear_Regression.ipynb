```python
def flip_n_times( times ):
    score = 0
    
    scoreArray = np.random.randint( 2, size=10) 
    score = sum(scoreArray)
    return score
```

```python
def flip_n_coins( num, times ):
    distribution = []
    for i in range( num ):
        score = flip_n_times( times )
        distribution.append( score )
    return distribution
```

```python
def get_v_min( num, times ):
    distribution = flip_n_coins( num, times )
    print(distribution)
    return min(distribution)
```

```python
def repeat_exp( repeat_time ):
    
    num = 1000
    times = 10
    
    sum_min = 0
    for i in range( repeat_time ):
        sum_min += get_v_min( num , times )

    avg_min = sum_min / repeat_time
    
    return avg_min
```

## Linear Regression

```python
import numpy as np
import random
```

```python
class Point:
    x = 0
    y = 0 
    def __init__(self,x,y):
        self.x = x
        self.y = y
```

```python
class Line:
    p1 = None
    p2 = None
    y_int = 0
    slope = 0
    
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        
        dx = p2.x - p1.x
        dy = p2.y - p1.y
    
        # check if the line is vertical
        if dx == 0 :
            self.is_vertical = True
            self.slope = None 
        
        else:
            # calculate slope
            self.is_vertical = False
            self.slope = dy/dx
        
        # derive the formula for the line in terms of x an y
        # (y - y1) = m(x - x1) where (x1,y1) is coord of a point and m is the slope
        # General Form y = mx + b, we know m so only b is needed
        # we call it y_int since it is the y interception
        
        self.y_int = p1.y - self.slope * p1.x
          
    
    def mapping(self, p):
        # to check which side is p at and thus decide its sign
        
        # if line is not vertical, compare the y values
        if self.is_vertical == False:
            line_y_value = self.slope*p.x + self.y_int
            diff = p.y - line_y_value
            
        # if line is vertical, compare the x values
        else:
            line_x_value = self.p1.x
            diff = p.x - line_x_value
        
        return np.sign(diff)        
            
```

```python
class Linear_Regression:
    

    
    def construct_X( self, points ):
        X = [ [ 1, point.x, point.y ] for point in points ]
        return X
    
    def calc_X_pseudo_inverse( self, X ):   
        # X(p.i.) = ( (X^T * X)^-1 X^T )
        trans = np.transpose( X )
        XT_X = trans.dot(X)
        inv = np.linalg.inv(XT_X)
        pseudo_inv = inv.dot(trans)
    
        return pseudo_inv    
```

```python
class Run_LR():
    
    num_of_sample = 0
    points = []
    target = None
    X = []
    Y = []
    lr = None
    weight = [0,0,0]
    
    def __init__( self, n ):
        self.num_of_sample = n
 
    def set_up(self):
        self.points = []
        for i in range(self.num_of_sample):
            x = random.uniform(-1, 1)
            y = random.uniform(-1, 1)
            self.points.append(Point(x, y))
        
        line_point_1_x = random.uniform(-1,1)
        line_point_1_y = random.uniform(-1,1)    
        line_point_1 = Point( line_point_1_x, line_point_1_y )
        
        line_point_2_x = random.uniform(-1,1) 
        line_point_2_y = random.uniform(-1,1)
        line_point_2 = Point( line_point_2_x, line_point_2_y )
        
        ## target function
        self.target = Line(line_point_1, line_point_2) 
    
    def train(self):
        self.Y = [[self.target.mapping(point)] for point in self.points]
        
        ## w = pseudoX * Y
        self.lr = Linear_Regression()
        self.X = self.lr.construct_X( self.points )
        pseudo_X = self.lr.calc_X_pseudo_inverse( self.X )
        self.weight = pseudo_X.dot( self.Y )
        
        
        return weight
    
    def test_E_in(self):
        
        e_sum = 0

        for p in self.points:
            
            x = [1, p.x, p.y]
            wT = self.weight.T 
            prediction = np.sign(wT.dot(x))
            actual_result = self.target.mapping(p)
           
            if prediction != actual_result:
                e_sum += 1

        
        e_in = e_sum / self.num_of_sample
        return e_in
    
    
    def test_E_out(self, testnum ):
        e_sum = 0
        for i in range( testnum ):
            p = Point(random.uniform(-1, 1),random.uniform(-1, 1))
            x = [1,p.x, p.y]
            wT = self.weight.T 
            prediction = np.sign(wT.dot(x))
            actual_result = self.target.mapping(p)
            if prediction != actual_result:
                    e_sum += 1


        e_out = e_sum / testnum
        return e_out
       
            
             
            
            
        

        
        
            


        
```

```python
## test E_in

E_in = []   
for i in range(1000):

    E_in = []
    test = Run_LR(100)
    test.set_up()
    test.train()
    E_in.append( test.test_E_in() )
print( float(np.average(E_in)))

print(test.test_E_out(1000))


    


        
    
```

```python
class Perceptron_Learning_Algorithm:
    def __init__(self,weight):
        self.weight = weight
    
    def predict(self, x):
        # x is a 2-D vector
        # vec_x = [ 1, x[1], x[2] ] x_0 is the artificial index
        vec_x = [1, x.x, x.y]
        dot_product = np.dot(vec_x, self.weight)
        return np.sign(dot_product)
    
    def train(self, x, y ): # x and y here refer to the domain and codomain of the function
        
        outcome = self.predict(x) ## yields 1/-1       
        if outcome != y:
            # update weight
            vec_x = [1, x.x, x.y]
            self.weight += np.multiply(y, vec_x )
        return outcome == y        
```

```python
class Run_PLA:
    def __init__(self, num_of_sample, weight):
        self.num_of_sample = num_of_sample
        ## generate the set of points
        self.points = test.points
        ## generate the line
        
        self.target = test.target
        self.PLA = Perceptron_Learning_Algorithm(weight)
    
    def test_agree( self, point ):
        
        cur_prediction = self.PLA.predict( point )
        actual_value = self.target.mapping( point )
        return cur_prediction == actual_value
    
    def test_convergence(self):
        iters = 0
        testidx = 0
        while True:
            
            actual_value = self.target.mapping(self.points[testidx])
            success = self.PLA.train(self.points[testidx], actual_value )
            
            if success:
                allgood = True
                
                for i in range(self.num_of_sample):
                    agree = self.test_agree(self.points[i])
                    if not agree:
                        allgood = False
                        break
                if allgood:
                    break
            else:
                iters = iters + 1
            testidx = int(np.random.uniform(0, self.num_of_sample-0.5))  
        return iters

    
    
    def disagree_probability(self):
            
            n_disagree = 0
            for x in range(1000):
                testpoint= Point(random.uniform(-1.0,1.0),random.uniform(-1.0,1.0))
                agree = self.test_agree(testpoint)
                if not agree:
                    n_disagree += 1

            prob = float(n_disagree)/1000.0
            return prob

            
            
```

```python
def main(test_dim, numtests, weight):
    converge_iters = []

    
    for a in range(numtests):
        cur_test = Run_PLA(test_dim, weight)
        cur_conv = cur_test.test_convergence()
        converge_iters.append(cur_conv)
        
    avg_iters = np.average(converge_iters)
    prefix_str = "For N = " + str(test_dim) + ", "
    print(prefix_str + "convergence took on average " + str(avg_iters) + " iterations")
```

```python
w = []
for i in range(3):
    test = Run_LR(100)
    test.set_up()
    test.train()
    w.append(float(test.weight[i]))
print(w)
main( 10, 1000, [0,0,0])
main( 10, 1000, w)
```
